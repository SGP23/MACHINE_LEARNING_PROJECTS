#Data Cleaning

```python
data=data.drop(['Address','Date','Postcode','YearBuilt','Lattitude','Longtitude'],axis=1)
```

```python
data.isnull().sum()
```

```python
data[['Propertycount','Distance',"Bedroom2",'Bathroom','Car']]=data[['Propertycount','Distance',"Bedroom2",'Bathroom','Car']].fillna(0)
```

```python
data[['Landsize']]=data[['Landsize']].fillna(data.Landsize.mean())
```

```python
data[['BuildingArea']]=data[['BuildingArea']].fillna(data.BuildingArea.mean())
```

```python
data["Method"]
```

```python
data['Method'].unique()
```

```python
data.dropna(inplace=True)
```

```python
data.shape
```

#Encoding Categorical Variables

```python
data=pd.get_dummies(data,drop_first=True)
```

```python
data
```

```python
data.shape
```

#Splitting Features and Target

```python
x=data.drop('Price',axis=1)
y=data['Price']
```

```python
x.head()
```

```python
x.tail()
```

#Model Training & Testing

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=10)
```

```python
x_train.shape
```

```python
x_test.shape
```

#Linear Regression Model

```python
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()
```

```python
model.fit(x_train,y_train)
```

#Model Prediction & Metrics

```python
model.score(x_test,y_test)
```

```python

model.score(x_train,y_train)
```

```python
model.predict(x)
```

```python

r2_score(y,model.predict(x))
```

```python

root_mean_squared_error(y,model.predict(x))
```

```python
mean_absolute_error(y,model.predict(x))
```

```python
mean_squared_error(y,model.predict(x))**(1/2)
```

#Regularized Models: Ridge, Lasso

```python
ridge_data= Ridge(alpha=50, max_iter=100,)
ridge_data.fit(x_train, y_train)
```

```python
ridge_data.score(x_test, y_test)
```

```python
ridge_data.score(x_train, y_train)
```

```python
lasso_data = Lasso(alpha=50, max_iter=100)
lasso_data.fit(x_train, y_train)
```

```python
lasso_data.score(x_train,y_train)
```

```python
lasso_data.score(x_test,y_test)
```
